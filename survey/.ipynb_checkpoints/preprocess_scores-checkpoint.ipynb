```python
# Preprocess Chatbot Survey Responses → Prompt × Condition × Scoreimport reimport jsonimport numpy as npimport pandas as pdfrom pathlib import PathSURVEY_FILE = Path("Chatbot Rating - Anwers bot masked.csv")BLOCKMAP_FILE = Path("surveyblock_map.csv")TASKS_FILE = Path("tasks_data.json")OUTPUT_FILE = Path("processed_scores.csv")# Load survey (first column is participant ID; we drop it)survey_df = pd.read_csv(SURVEY_FILE, encoding="utf-8")pid_col = survey_df.columns[0]# Load block map and rename matching columnblockmap = pd.read_csv(BLOCKMAP_FILE, encoding="utf-8")if "column" not in blockmap.columns:    survey_cols = set(survey_df.columns.astype(str))    overlap = {c: blockmap[c].astype(str).isin(survey_cols).mean() for c in blockmap.columns}    best = max(overlap, key=overlap.get)    blockmap = blockmap.rename(columns={best: "column"})def pick_first(d, keys, default=np.nan):    for k in keys:        if k in d:            return d[k]    return defaultbm = blockmap.copy()if "construct" not in bm.columns:    candidate = pick_first(bm, ["metric","Metric","METRIC"])    bm["construct"] = candidate if isinstance(candidate, pd.Series) else np.nanif "reverse" not in bm.columns:    cand = pick_first(bm, ["reverse_scored","Reverse","rev","REV"])    bm["reverse"] = cand if isinstance(cand, pd.Series) else 0if "is_filler" not in bm.columns:    cand = pick_first(bm, ["filler","is filler","isFiller","FILLER_FLAG"])    bm["is_filler"] = cand if isinstance(cand, pd.Series) else 0if "is_attention" not in bm.columns:    cand = pick_first(bm, ["attention","attn","ATT_FLAG","att_check_flag"])    bm["is_attention"] = cand if isinstance(cand, pd.Series) else 0if "Prompt_Id" not in bm.columns:    cand = pick_first(bm, ["prompt_id","PromptID","Prompt","prompt","PROMPT_ID"])    bm["Prompt_Id"] = cand if isinstance(cand, pd.Series) else np.nanif "condition" not in bm.columns:    cand = pick_first(bm, ["Condition","COND","cond"])    bm["condition"] = cand if isinstance(cand, pd.Series) else np.nanbm["column"] = bm["column"].astype(str)for c in ["construct","Prompt_Id","condition"]:    bm[c] = bm[c].astype(str).str.strip()bm["construct"] = bm["construct"].str.upper()bm["Prompt_Id"] = bm["Prompt_Id"].str.upper()bm["condition"] = bm["condition"].str.upper()for c in ["reverse","is_filler","is_attention"]:    bm[c] = pd.to_numeric(bm[c], errors="coerce").fillna(0).astype(int)long_df = survey_df.melt(id_vars=[pid_col], var_name="column", value_name="raw_score")long_df["raw_score"] = pd.to_numeric(long_df["raw_score"], errors="coerce")long_df["column"] = long_df["column"].astype(str)def infer_condition(s):    m = re.match(r"^\s*([AB])(?:[\/_\-\s]|$)", str(s).upper())    return m.group(1) if m else np.nandef infer_prompt(s):    m = re.search(r"(P\d{1,3})", str(s), flags=re.IGNORECASE)    return m.group(1).upper() if m else np.nanbm["condition"] = bm["condition"].replace({"NAN": np.nan})bm["Prompt_Id"] = bm["Prompt_Id"].replace({"NAN": np.nan})bm["cond_fallback"] = bm["column"].apply(infer_condition)bm["pid_fallback"] = bm["column"].apply(infer_prompt)bm.loc[bm["condition"].isna(), "condition"] = bm.loc[bm["condition"].isna(), "cond_fallback"]bm.loc[bm["Prompt_Id"].isna(), "Prompt_Id"] = bm.loc[bm["Prompt_Id"].isna(), "pid_fallback"]bm = bm.drop(columns=["cond_fallback","pid_fallback"])df = long_df.merge(bm, on="column", how="left")df.loc[df["condition"].isna(), "condition"] = df.loc[df["condition"].isna(), "column"].apply(infer_condition)df.loc[df["Prompt_Id"].isna(), "Prompt_Id"] = df.loc[df["Prompt_Id"].isna(), "column"].apply(infer_prompt)df = df[df["construct"].notna()].copy()df["construct"] = df["construct"].str.upper().str.strip()df["condition"] = df["condition"].astype(str).str.upper().str.strip()df["Prompt_Id"] = df["Prompt_Id"].astype(str).str.upper().str.strip()def reverse_score(x):    return np.nan if pd.isna(x) else 6 - xdf["score"] = df["raw_score"]df.loc[df["reverse"] == 1, "score"] = df.loc[df["reverse"] == 1, "score"].apply(reverse_score)if "is_filler" in df.columns:    df = df[df["is_filler"] != 1]df = df[df["construct"] != "FILLER"]if (df["construct"] == "ATT_CHECK").any():    att = (df[df["construct"] == "ATT_CHECK"]           .groupby(pid_col)["score"].apply(list)           .reset_index(name="att_scores"))    def pass_att(xs):        xs = [v for v in xs if pd.notna(v)]        return int(len(xs) > 0 and all(v == 5 for v in xs))    att["pass_attention"] = att["att_scores"].apply(pass_att)    df = df.merge(att[[pid_col,"pass_attention"]], on=pid_col, how="left")    df = df[df["pass_attention"] == 1]targets = {"PI","CSAT","HL","CMP"}df = df[df["construct"].isin(targets)]scored = (df.groupby(["Prompt_Id","condition","construct"])["score"]            .mean()            .reset_index())wide = (scored.pivot_table(index=["Prompt_Id","condition"],                           columns="construct", values="score")                .reset_index())wide.columns.name = Nonewith open(TASKS_FILE, "r", encoding="utf-8") as f:    tasks = json.load(f)tasks_df = pd.json_normalize(tasks)for c in ["meta.prompt_id","meta.condition","meta.reply_id","data.bot_mask","data.reply_text"]:    if c not in tasks_df.columns:        tasks_df[c] = np.nantasks_df["Prompt_Id"] = tasks_df["meta.prompt_id"].astype(str).str.upper()tasks_df["condition"] = tasks_df["meta.condition"].astype(str).str.upper()tasks_df["reply_id"]  = tasks_df["meta.reply_id"].astype(str)tasks_df["bot_mask"]  = tasks_df["data.bot_mask"]tasks_df["reply_text"] = tasks_df["data.reply_text"]tasks_df = tasks_df[["Prompt_Id","condition","reply_id","bot_mask","reply_text"]].drop_duplicates()out = wide.merge(tasks_df, on=["Prompt_Id","condition"], how="left")out.to_csv(OUTPUT_FILE, index=False, encoding="utf-8")print(f"Saved → {OUTPUT_FILE}")out.head()
```
